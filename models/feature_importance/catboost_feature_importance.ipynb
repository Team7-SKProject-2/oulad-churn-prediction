# CatBoost 예측 영향 Feature 상위 N개 추출

`catboost_feature_importance.py`는 최종 후보 CatBoost를 GroupKFold 기준 Train 학생으로 학습한 뒤, Validation 학생 표본에서 평균 절대 SHAP 값을 계산합니다.

- `CUMULATIVE_IMPORTANCE_MIN`, `MAX`, `TARGET`: 누적 영향력 70~80% 범위에서 75%에 가장 가까운 상위 Feature를 자동 선택
- `SHAP_SAMPLE_SIZE`: SHAP 계산에 사용할 Validation 행 수
- 결과: 전체 중요도 CSV, 누적 목표에 도달하는 최소 N개 Feature CSV, 막대그래프 PNG

SHAP 중요도는 **예측에 미친 영향의 크기**이며 인과관계나 이탈 원인을 단정하는 값은 아닙니다.


```python
from pathlib import Path
import runpy

current = Path.cwd().resolve()
script_path = current / 'catboost_feature_importance.py'
if not script_path.exists():
    script_path = current / 'models' / 'feature_importance' / 'catboost_feature_importance.py'

if not script_path.exists():
    raise FileNotFoundError('models/feature_importance/catboost_feature_importance.py 파일을 찾을 수 없습니다.')

runpy.run_path(str(script_path), run_name='__main__')
```
