## courses(강의) 데이터 체크(1차 전처리)


```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve()

# 노트북에서 실행할 경우 프로젝트 최상위 폴더로 이동
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

COURSES_PATH = PROJECT_ROOT / "data" / "raw" / "courses.csv"

print("프로젝트 위치:", PROJECT_ROOT)
print("courses 존재:", COURSES_PATH.exists())
```

```python
courses = pd.read_csv(COURSES_PATH)

print("courses 크기:", courses.shape)
display(courses.head())
```

```python
print("===== courses 정보 =====")
courses.info()

print("\ncourses 컬럼별 결측/공백/'?' 개수:")
found_missing = False
for col in courses.columns:
    none_count = (
        courses[col].isnull()
        | (courses[col].astype(str).str.strip() == "")
        | (courses[col].astype(str).values == "?")
    ).sum()
    if none_count > 0:
        found_missing = True
        print(f"{col} : {none_count}")

if not found_missing:
    print("결측치 없음")
```

```python
COURSE_KEYS = ["code_module", "code_presentation"]

print("핵심 컬럼 결측치:")
print(courses[COURSE_KEYS].isna().sum())

print("\n키 중복 행 수:", courses.duplicated(subset=COURSE_KEYS).sum())

print("\nmodule_presentation_length 분포:")
print(courses["module_presentation_length"].describe())
```

결측치와 키 중복이 없고, 강좌 운영 기간(module_presentation_length)도 이상치 없이 정상 범위이므로
별도의 정제 없이 원본을 그대로 정제본으로 사용한다.


```python
courses_processed = courses.copy()

print("정제본 크기:", courses_processed.shape)
print("정제본 결측치:")
print(courses_processed.isna().sum())
```

```python
# 저장 경로
INTERIM_DIR = PROJECT_ROOT / "data" / "interim"
INTERIM_DIR.mkdir(parents=True, exist_ok=True)

COURSES_OUTPUT_PATH = INTERIM_DIR / "courses_processed.csv"

courses_processed.to_csv(COURSES_OUTPUT_PATH, index=False)

print("저장 완료")
print(
    COURSES_OUTPUT_PATH.name,
    f"{COURSES_OUTPUT_PATH.stat().st_size / 1024**2:.2f} MB"
)
```
